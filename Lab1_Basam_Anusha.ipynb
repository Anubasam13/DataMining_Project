# QUESTION 4
The objective of the four labs you are completing for this course is to determine the time it takes to clear the scene after each accident in the city of Denton. There are four columns that collect data and time-related information, including the time of the accident and the time when the scene was cleared. The first step involves combining the â€œDateâ€ and â€œTimeâ€ columns into a single column in order to compute the duration it will take to clear the scene. The difference between â€œTime Scene Clearedâ€ and â€ Accident Timeâ€ in minutes in the â€œClearance Time (Minutes)â€. Note that to correctly compute the â€œClearance Time (Minutes)â€ you should consider both data and time columns. After completing this tasks, enter the average â€œClearance Time (Minutes)â€ for all the accidents in this data frame. Enter the answer with two digits after the decimal point.

````python
# Coverting date and time format to datetime format
df["Crash_date_time"] = pd.to_datetime(df["Crash Date"] + " " + df["Crash Time"].astype(str).str.zfill(4), format = "%Y-%m-%d %H%M")
df["Scene Cleared DateTime"] = pd.to_datetime(df["Date Scene Cleared"] + " " + df["Time Scene Cleared"].astype(str).str.zfill(4), format = "%Y-%m-%d %H%M") 

# Calculate clearance time in minutes
df["Clearance_time_minutes"] = (df["Scene Cleared DateTime"] - df["Crash_date_time"]).dt.total_seconds() / 60

# Calculate average clearance time

avg_clearance_time = df["Clearance_time_minutes"] .mean()

# output

print(f"Average Clearance Time in Minutes : {avg_clearance_time:.2f}")
````

# Question 19
1.	We need to create a new column called "Time of Day" based on the values in the "Crash Time" column. The "Time of Day" column will contain the values "Morning Rush Hour", "Evening Rush Hour", and "Other". The time ranges for the morning and evening rush hours are as follows:

  o	Morning Rush Hour: 7:00 AM to 9:00 AM

  o	Evening Rush Hour: 4:00 PM to 6:00 PM

  o	Other: All other times

The next step involves transforming the "Time of Day" column appropriately. Choose and apply the correct data transformation method for this column:

a.	Map values as follows: "Morning Rush Hour" to -1, "Other" to 0, and "Evening Rush Hour" to 1.

b.	Map values as follows: "Other" to 0, "Morning Rush Hour" to 1, "Evening Rush Hour" to 2.

c.	Replace it with a one-hot vector of length 3.

d.	Map the values as follows: "Morning Rush Hour" to 0, "Evening Rush Hour" to 1, and "Other" to 2.


````python
# Convert 'Crash Time' to a string format with leading zeros
df ["Crash Time"] = df["Crash Time"].astype(str).str.zfill(4)

# Define a function to categorize time of day
def categorize_time_of_day (time_str):
    hour = int(time_str[:2]) # Get the hour
    minute = int(time_str[:2]) # Get the minute

    if 7 <= hour < 9: # Morning Rush Hour (7 AM to 9 AM)
        return "Morning Rush Hour"
    elif 16 <= hour < 18: # Evening Rush Hour (4 PM to 6 PM)
        return "Evening Rush Hour" 
    else:
        return "Other"
# Apply the function to create the 'Time of Day' column using Replace it with a one-hot vector of length 3.
df["Time of Day"] = df["Crash Time"].apply(categorize_time_of_day)
df["Time of Day"].value_counts()
````

When dealing with the "Surface Condition" column, we will use the same approach as previously discussed. Any categories with fewer than 50 instances will be grouped under a new category called "Other." Additionally, we will remove the numerical code preceding the condition description. For example, for "1 - DRY," we will only consider "DRY." b. Replace it with a one-hot vector of length 3.

````python
# calling the column sum
fre = df["Surface Condition"].value_counts()
fre
# Any categories with fewer than 50 instances will be grouped under a new category called "Other."
less_fre = fre[fre < 50].index
df.loc[df["Surface Condition"].isin(less_fre), "Surface Condition"] = "Other"

# Counting the updated frequency
upd_freq = df["Surface Condition"].value_counts()

# Print the updated frequency
print(upd_freq)
df[df["Surface Condition"] == "Other"]["Surface Condition"].value_counts()

# Removing the numerical code preceding the condition description
surface_map = {
    "1 - DRY": "DRY",
    "2 - WET": "WET",
    "Other": "Other"}

# Applying mapping
df["Surface Condition"] = df["Surface Condition"].map(surface_map)
df["Surface Condition"]
````

# Question 21
Given our previous discussion, we need to apply the following logic to the â€œWeather Conditionâ€ column: if a category has fewer than 50 instances, it will be grouped under the category â€œOTHER.â€ Additionally, we should remove the numerical code from the condition description. For instance, for â€œ1 - CLEAR,â€ we should only consider â€œCLEAR.â€ With this in mind, what would be the most effective data transformation technique for this column?

a. Map values according to their frequencies in that column to ordinal values 0 to 3.
    
b. Replace it with a one-hot vector of length 8.

c. Map values according to their frequencies in that column to ordinal values 1 to 8.
    
d. Replace it with a one-hot vector of length 4


````python
# Calling the column
df["Weather Condition"]

# Checking fewer than 50 instances
fre_count = df["Weather Condition"].value_counts()
fre_count

# Grouping fewer than 50 instances to "Other"
low_fre = fre_count[fre_count < 50].index
df.loc[df["Weather Condition"].isin(low_fre), "Weather Condition"] = "Other"

# Counting the updated frequency
updated_freq = df["Weather Condition"].value_counts()

# Print the updated frequency
print(updated_freq)
df[df["Weather Condition"] == "Other"]["Weather Condition"].value_counts()

# Removing the numerical code preceding the condition description
weather_map = {
    "1 - CLEAR": "CLEAR",
    "2 - CLOUDY": "CLOUDY",
    "3 - RAIN": "RAIN", 
    "Other": "Other"}

# Applying weather mapping
df["Weather Condition"] = df["Weather Condition"].map(weather_map)
df["Weather Condition"]

# Map values according to their frequencies in that column to ordinal values 0 to 3
ordinal_map = {
    "Other": 0,
    "CLEAR": 1,
    "CLOUDY": 2,
    "RAIN": 3}

df["Weather Condition"] = df["Weather Condition"].map(ordinal_map)
df["Weather Condition"]
````
